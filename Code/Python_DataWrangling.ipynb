## 2. Merge Dataset to Prepare for Analysis:

```python
n_features.rename(columns={'IsHoliday' : "Holiday"}, inplace=True)
n_features.head()
```

```python
n_sales.rename(columns={'IsHoliday' : "Holiday"}, inplace=True)
n_sales.head()
```

```python
s_sales = pd.merge(n_stores, n_sales, on='Store')
s_sales
```

```python
retail = pd.merge(s_sales, n_features)
retail.head()
```

```python
retail.groupby('Store')['Weekly_Sales'].sum()
```

```python
retail[["Year", "Month", "Day"]] = retail["Date"].astype(str).str.split("-", expand = True)
```
